```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load dataset
housing = fetch_california_housing()

# Create DataFrame
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['Price'] = housing.target  # Add target variable

# Save raw data
df.to_csv('../data/raw/california_housing.csv', index=False)

# Explore the data
print(df.info())
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.savefig('../data/raw/correlation_matrix.png')

# Distribution of target variable
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True)
plt.title('Housing Price Distribution')
plt.xlabel('Price')
plt.savefig('../data/raw/price_distribution.png')

# Pairplot for key features
sns.pairplot(df[['MedInc', 'HouseAge', 'AveRooms', 'Price']])
plt.savefig('../data/raw/pairplot.png')
```
